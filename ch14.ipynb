```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
from collections import defaultdict

def get_counts2(sequence):
    counts = defaultdict(int) # values will initialize to 0
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
weather_df = pd.DataFrame(np.random.rand(10,2)*5,
                          index=pd.date_range(start="2021-01-01", periods=10),
                          columns=["Tokyo", "Beijing"])

def rain_condition(v):
    if v < 1.75:
        return "Dry"
    elif v < 2.75:
        return "Rain"
    return "Heavy Rain"
```

```python

def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.format(rain_condition)
    styler.format_index(lambda v: v.strftime("%A"))
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
    return styler

weather_df
weather_df.loc["2021-01-04":"2021-01-08"].style.pipe(make_pretty)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': np.random.randn(5)*100,
    'B': np.random.randint(1, 100, 5),
    'C': ['X', 'Y', np.nan, 'X', 'Z']
})

# 设置全局样式
styled = df.style \
    .set_properties(**{'background-color': '#f8f8f8', 'border': '1px solid #ccc'}) \
    .format('{:.1f}', subset=['A', 'B']) \
    .format({'C': lambda x: x or "NULL"}) \
    .highlight_null('yellow')  # 高亮缺失值
styled
```

```python
# 列值颜色映射
styled = df.style \
    .background_gradient(subset=['A'], cmap='Reds') \
    .bar(subset=['B'], color='#5fba7d', align='zero')  # 条形图

# 自定义条件格式
def highlight_high(val):
    color = 'red' if val > 80 else 'black'
    return f'color: {color}'

styled.applymap(highlight_high, subset=['B'])
```

```python
# 表头样式
headers = {
    'selector': 'th',
    'props': [('background-color', '#40466e'), ('color', 'white')]
}

# 隔行换色
row_styles = {
    'selector': 'tr:nth-child(even)',
    'props': [('background-color', '#f2f2f2')]
}

styled.set_table_styles([headers, row_styles])
```
